#**import** **libraries**

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
df=sns.load_dataset("planets")
df.head()
df.info()
df.describe()
```

# **count plot**

```python
sns.countplot(data=df, y="method"); plt.title("Planet Discovery Methods"); plt.show()
```

# **histogram**

```python
plt.figure(figsize=(8,5))
sns.histplot(
    data=df,
    x="orbital_period",
    bins=30,
    kde=True
)
plt.title("distribution of orbital period")
plt.xlabel("orbital period(days)")
plt.ylabel("number of planets")
plt.show()
```

# box **plot**

```python
sns.boxplot(data=df, x="method", y="orbital_period"); plt.xticks(rotation=90); plt.show()
```

# **scatterplot**

```python
sns.scatterplot(data=df, x="distance", y="mass"); plt.title("Distance vs Mass"); plt.show()
```

# bar **plot**

```python
sns.barplot(data=df, x="method", y="mass"); plt.xticks(rotation=90); plt.show()
```

# line **plot**

```python
sns.lineplot(data=df, x="year", y="mass"); plt.title("Year vs Mass"); plt.show()
```

# correlation **heatmap**

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm"); plt.show()
```

# pair **plot**

```python
sns.pairplot(df[["orbital_period","mass","distance","year"]].dropna()); plt.show()
```

# KDE plot-**planet** mass

```python
plt.figure(figsize=(7,5))

sns.kdeplot(
    data=df,
    x="mass",
    fill=True
)

plt.title("Density Distribution of Planet Mass")
plt.xlabel("Mass")

plt.show()
```

# violin **plot**

```python
sns.violinplot(data=df, x="method", y="orbital_period"); plt.xticks(rotation=90); plt.show()
```

# swarm **plot**

```python
plt.figure(figsize=(12,6))

sns.swarmplot(data=df, x="method", y="mass", size=2)

plt.xticks(rotation=90)
plt.title("Planet Mass by Discovery Method")
plt.show()
```

# strip **plot**

```python
plt.figure(figsize=(12,6))

sns.stripplot(data=df, x="method", y="mass", jitter=True)

plt.xticks(rotation=90)
plt.title("Planet Mass by Discovery Method")
plt.show()
```
